# Lab Exercise: Payroll Tax Functions

lab makes three different functions that calculates taxes for 
social security
medicare
and federal income tax 
all 3 are seperated with 3 samples


```python
def socSecTax(grossPay):
    #its like 6%
    return grossPay *0.062
```

```python
def medTax(grossPay):
    #oh nice its even less
    return grossPay * 0.0145
```

```python
def fedTax(grossPay, evilMoneyStealingCode):
    if evilMoneyStealingCode == 0:
        rate = 0.23
    elif evilMoneyStealingCode == 1:
        rate = 0.21
    elif evilMoneyStealingCode == 2:
        rate = 0.195
    elif evilMoneyStealingCode == 3:
        rate = 0.185
    else: #4 moment
        excess = evilMoneyStealingCode - 4
        rate = 0.18 - (excess * 0.005) #Maybe? I hope this works
        if rate < 0: # not needed but like, dont want to send someone infinite money that you then have to arrest them for
            rate = 0
    return grossPay * rate
```

```python
gross1 = 750
code1 = 0

ss_tax1 = socSecTax(gross1)
med_tax1 = medTax(gross1)
fed_tax1 = fedTax(gross1, code1)

print(f"person 1 - gross: ${gross1:.2f}, Code: {code1}")
print(f"pyramid scheme tax:${ss_tax1:.2f}")
print(f"medicare tax: ${med_tax1:.2f}")
print(f"federal tax: ${fed_tax1:.2f}")
print(f"total deductions(sad!): ${ss_tax1 + med_tax1 + fed_tax1:.2f}")
print(f"net pay: ${gross1 - (ss_tax1 + med_tax1 + fed_tax1):.2f}")
```
