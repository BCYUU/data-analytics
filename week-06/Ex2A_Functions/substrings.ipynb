## Lab 1. Working with names – parsing strings

```python
test_name = 'Joe Biden'
```

```python
def parse_and_display(name):
    for letter in range(len(name)):
        if  name[letter] == " ":
            first = name[:letter]
            last = name[letter+1:]
            print(name)
            print(first)
            print(last)

```

```python
parse_and_display("Joe Biden")
```

```python
parse_and_display(test_name) 
```

```python
test_list = ['Grace Flores', 'JB Oinonen', 'Ken Penn']
```

```python
for i in test_list:
    parse_and_display(i)
```
