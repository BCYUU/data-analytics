Examples:

ACME:123-L supplied by Acme, part #123, size large

DI:12345-M supplied by DI, part #12345, size medium

ACE:1-12 supplied by Ace, part #1, size 12

```python
partList = ["ACME:123-L",
"DI:12345-M",
"ACE:1-12"]
```

```python
def getScode(pcode):
    if ":" not in pcode:
        return ""
    return pcode.split(":")[0]
```

```python
def getPcode(pcode):
    if ":" not in pcode or '-' not in pcode:
        return ""
    
    afterSupplier = pcode.split(":",1)[1]
    return afterSupplier.split("-",1)[0]
```

```python
def getSize(pcode):
    if "-" not in pcode:
        return ""
    return pcode.split("-",1)[1]
```
