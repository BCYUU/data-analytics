CUSTOM FUNCTIONS


```python
def favThings(name, movie, music):
    return f"My name is {name}.\nMy favorite movie is {movie}.\nMy favorite musician is {music}."

print(favThings("Blake", "He Never Died", "Elvis Presley"))
```

```python
def whyImHere(purpose):
    return purpose

print(whyImHere("I joined the Year Up United Data Analyst program to gain practical data skills, because my buddy told me about it."))
```

```python
def favPLace(place, why):
    return f"One of my favorite places is my {place} because {why}."

print(favPLace("Grand Parents Home", "spent so many weekends at peace over here"))
```

```python
print(favPLace("Local Mountain", "I like hiking"))
```
