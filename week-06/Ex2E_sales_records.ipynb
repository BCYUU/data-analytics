### Exercise 2.E Using functions for analyzing data

This notebook analyzes a small sales dataset using Python functions. Each function takes the full dataset as its first argument.

```python
sales_data = [ 
    ('Laptop',     'East',  1200.00), 
    ('Phone',      'West',   850.00), 
    ('Tablet',     'East',   400.00), 
    ('Monitor',    'South',  650.00), 
    ('Keyboard',   'West',    75.00), 
    ('Laptop',     'South', 1150.00), 
    ('Phone',      'East',   900.00),
    ('Tablet',     'West',   425.00), 
    ('Monitor',    'East',   680.00), 
    ('Keyboard',   'South',   80.00), 
    ('Laptop',     'West',  1300.00), 
    ('Phone',      'West',   820.00), 
]
```

## Lab 1. Total sales by region


```python
def regTotal(data, region):
    return sum(amt for _, r, amt in data if r == region)
```

```python
print(regTotal(sales_data, 'East'))
print(regTotal(sales_data, 'West'))
print(regTotal(sales_data, 'South'))
```

## Lab 2. Count of sales by region

```python
def regCount(data, region):
    # counts how number of records in the region
    return sum(1 for _, r, _ in data if r == region)
```

```python
print(regCount(sales_data, 'East'))
print(regCount(sales_data, 'West'))
print(regCount(sales_data, 'South'))
```

## Lab 3.  Average sale amount by region

```python
def regAvg(data, region):
    total = regTotal(data, region)
    count = regCount(data, region)
    return total / count if count > 0 else 0 # imaginary solution for a / by zero situation
```

```python
print(regAvg(sales_data, "East"))
print(regAvg(sales_data, "West"))
print(regAvg(sales_data, "South"))
print(regAvg(sales_data, "North")) # for example
```

## Lab 4.  Filter by sale amount 

```python
def salesAbove(data, threshold):
    return [record for record in data if record[2] > threshold]

highSales = salesAbove(sales_data, 500)

for record in highSales:
    print(record)
```

## Lab 5.   Top and bottom sale

```python
def topSale(data):
    return max(data, key=lambda record: record[2])

def bottomSale(data):
    return min(data, key=lambda record: record[2])
```

```python
print("top sale", topSale(sales_data))
print("bottom sale", bottomSale(sales_data))
```

## Lab 6. Apply a discount

```python
def applyDiscount(data, discount):
    return list(map(lambda s: (s[0],s[1],s[2]*(1-discount)), data)) #thanks demand planning guy
```

```python
print("first 3 no discounts: ")
for record in sales_data[:3]:
    print(record)
    
```

```python
discountedData = applyDiscount(sales_data, 0.15)

print("\n now with the discount: ")
for record in discountedData[:3]:
    print(record)
```

## Lab 7. Unique regions 

```python
def uniReg(data):
    return sorted({region for _, region, _ in data})

print(uniReg(sales_data))
```

## Lab 8. Region summary 

```python
def regSum(data, region):
    regRec = [record for record in data if record[1] == region]
    total = regTotal(data, region)
    count = regCount(data, region)
    avg = regAvg(data, region)

    if count == 0:
        topProduct = "none"
        topAmount = 0.0
    else:
        topRecord = topSale(regRec)
        topProduct = topRecord[0]
        topAmount = topRecord[2]
    print(f"~~~~~~ {region} region ~~~~~~")
    print(f"records: {count}")
    print(f"total sales: ${total:,.2f}")
    print(f"average sale: ${avg:,.2f}")
    print(f"top sale: {topProduct} — ${topAmount:,.2f}\n")
```

```python
regSum(sales_data, "East")
regSum(sales_data, "West")
regSum(sales_data, "South")
regSum(sales_data, "North")
```

## Lab 9. Full summary  

```python
regions = uniReg(sales_data)

for region in regions:
    regSum(sales_data, region)
```
