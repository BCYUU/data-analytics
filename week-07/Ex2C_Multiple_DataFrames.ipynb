# Exercise 2.C Working with multiple DataFrames

```python
import numpy as np
import pandas as pd
```

## Basic Aggregations

```python
data = {
    'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North'],
    'Units_Sold': [10, 15, 20, 5, 10, 15, 10, 5, 15],
    'Price': [100, 200, 150, 100, 200, 150, 100, 200, 150]
}
dfSales = pd.DataFrame(data)


dfSales['Revenue'] = dfSales['Units_Sold'] * dfSales['Price']

totalUnits = dfSales['Units_Sold'].sum()
totalRevenue = dfSales['Revenue'].sum()
print(f"total units sold: {totalUnits}")
print(f"total revenue: {totalRevenue}")

avgUnits = dfSales['Units_Sold'].mean()
avgRevenue = dfSales['Revenue'].mean()
print(f"average units sold: {avgUnits:.2f}")
print(f"average revenue: {avgRevenue:.2f}")

```

## Grouping Data


```python
productGroup = dfSales.groupby('Product')[['Units_Sold', 'Revenue']].sum()
print("total by product:\n", productGroup)
```

```python
regionGroup = dfSales.groupby('Region')[['Units_Sold', 'Revenue']].mean()
print("Average by Region:\n", regionGroup)
```

```python
productRegionGroup = dfSales.groupby(['Product', 'Region'])[['Units_Sold', 'Revenue']].sum()
print("sum by Product and Region:\n", productRegionGroup)
```

## Sorting and Ranking Data


```python
dfSalesSortedUnits = dfSales.sort_values('Units_Sold', ascending=True)
print("Sorted by Units_Sold (ascending):\n", dfSalesSortedUnits)
```

```python
dfSalesSortedRevenue = dfSales.sort_values('Revenue', ascending=False)
print("Sorted by Revenue (descending):\n", dfSalesSortedRevenue)
```

```python
dfSales['Revenue_Rank'] = dfSales['Revenue'].rank(ascending=False, method='dense').astype(int)
dfSales
```

## Merging DataFrames

```python
productData = {'Product': ['A', 'B', 'C'],
               'Category': ['Electronics', 'Furniture', 'Appliances']}
dfProducts = pd.DataFrame(productData)
```

```python
dfProducts
```

```python
dfMerged = pd.merge(dfSales, dfProducts, on='Product')
dfMerged
```
