## Cumulative Operations

```python

print("r_4:\n", r_4)
print("sum of all elements:", np.sum(r_4))
print("sum along (columns):", np.sum(r_4, axis=0))
print("sum along (rows):   ", np.sum(r_4, axis=1))
print("cumulative sum (flattened):", np.cumsum(r_4))
print("cumulative sum along axis 0:\n", np.cumsum(r_4, axis=0))
```

## Grocery Problem

```python

bag1 = np.array([2, 3, 4, 0]) # 2 veg, 3 fruit, 4 snacks, 0 beverages
bag2 = np.array([4, 0, 1, 2]) # 4 veg, 0 fruit, 1 snack, 2 beverages

print("bag1\n", bag1)
print("bag2\n", bag2)
```

```python
totalBag1 = np.sum(bag1)
totalBag2 = np.sum(bag2)
print(f"The first bag has {totalBag1} items, and the second bag has {totalBag2} items.")
```

```python
totalPerCategory = bag1 + bag2
print(f"There are {totalPerCategory[0]} vegetables, {totalPerCategory[1]} fruits, {totalPerCategory[2]} snacks, and {totalPerCategory[3]} beverages in the bags.")
```

## Teamwork

```python
ali = np.array([5, 4, 6, 4, 2])
ben = np.array([0, 5, 5, 3, 4])
carla = np.array([3, 6, 0, 3, 4])

# array
allKnocks = np.array([ali, ben, carla])
print("All knocks:\n", allKnocks)
```

```python
# per person
personTotals = np.sum(allKnocks, axis=1)
# the group 
groupTotal = np.sum(allKnocks)

print("per person:", personTotals)
print("group total:", groupTotal)
```

```python
# rolling total by day
dailyTotal = np.sum(allKnocks, axis=0)      # shape (5,?)

estimatedVotes = dailyTotal // 2            # one in two will actually vote
print("doorknocks:", dailyTotal)
print("total expected votes per day:", estimatedVotes)
```
