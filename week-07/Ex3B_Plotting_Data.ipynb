# Exercise 3.B More plotting practice

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Advanced Plotting Techniques

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]
```

```python
plt.bar(products, sales, color='skyblue', width=0.5)
plt.title('Product Sales')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 100
plt.scatter(x, y, s=sizes, c=sizes, cmap='viridis', alpha=0.6)
plt.title('Scatter Plot with Colored Markers')
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(label='Size factor')
plt.show()
```

## Subplots

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

xLine = np.linspace(0, 10, 100)
yLine = np.sin(xLine)
ax1.plot(xLine, yLine, color='blue')
ax1.set_title('Sine Wave')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')

randomData = np.random.randn(1000)
ax2.hist(randomData, bins=20, color='green', edgecolor='black')
ax2.set_title('Histogram of Random Numbers')
ax2.set_xlabel('Value')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
