## Generating Random Choices from

```python

students = np.array(["Alice", "Bob", "Charlie", "David", "Ella", "Frank", 
"Grace", "Henry", "Ivy", "Jack", "Kara", "Liam"])
```

```python
callOn = np.random.choice(students)
print(callOn)
```

```python
breakoutGroups = np.random.choice(students, size=12, replace=False).reshape(3, 4)
print(breakoutGroups)
```

## Generating a Normal Distribution

```python
bellCurve = np.random.normal(loc=10, scale=5, size=(4, 5))
print(bellCurve)
```
