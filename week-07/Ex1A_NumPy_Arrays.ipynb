 ## Looping Through Arrays

```python
arrA = np.array([[11, 12, 13],[14, 15, 16]])
print(arrA)
```

```python
for x in arrA:
    print(x*2)
```

```python
for x in arrA:
    for y in x:
        print(y*2)
```

one loops over the rows of the array 
and the second one multipliues the individual elements

```python
for x in arrA:
    for y in x:
        loopResult = x*2
        print(loopResult, arrA.ndim, arrA.shape) 
```

so the first part multiplies the row and then the second part prints the row for each element in the row
